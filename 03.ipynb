# Tensorflow 세 가지 모델 작성


### Data Slicing

```python
#import, 함수 선언 부
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

#data slicing
data = np.loadtxt("data/wdbc_modified.csv", delimiter=",")
print(data.shape)

x_train = data[:, 0:30]
y_train = data[:, 30:]
x_new = x_train[15:25]

print(x_train.shape)
print(y_train.shape)
print(x_train[:10])
print(y_train[:10])
```

***
# Sequential Model
### Sequential Model 생성

```python
sqModel = keras.Sequential()
sqModel.add(layers.Dense(64, input_dim=30, activation="relu"))
sqModel.add(layers.Dense(32, activation="relu"))
sqModel.add(layers.Dense(1, activation="sigmoid"))

sqModel.summary()
```

### Sequential Model 학습 수행

```python
sqModel.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = sqModel.fit(x_train, y_train, epochs=500, batch_size=10)
```

### 알고 있는 y data에 대한 검증

```python
score = sqModel.evaluate(x_train, y_train, batch_size=10)
print(score)
```

### 새로운 데이터 예측

```python
y_predict = sqModel.predict(x_new)
print(y_predict)
print(y_train[15:25])
```

# Keras Functional API
### 모델 생성

```python
inputs = keras.Input(shape=(30,), name="my_input")
x = layers.Dense(64, activation="relu", name="layers")(inputs)
x = layers.Dense(32, activation="relu")(x)
outputs = layers.Dense(1, activation="sigmoid")(x)
funcModel = keras.Model(inputs=inputs, outputs=outputs)

funcModel.summary()
```

### 학습 수행

```python
funcModel.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = funcModel.fit(x_train, y_train, epochs=500, batch_size=10)
```

### y data 통해 검증

```python
#알고 있는 y data 통해 검증
score = funcModel.evaluate(x_train, y_train, batch_size=10)
print(score)
```

### 새로운 데이터 통해 예측

```python
y_predict = funcModel.predict(x_new)
print(y_predict)
print(y_train[15:25])
```

# Model Subclassing
### 모델 생성

```python
class MyModel(keras.Model):
    def __init__(self):
        super().__init__()
        self.layer1 = layers.Dense(64, input_dim=30, activation="relu")
        self.layer2 = layers.Dense(32, activation="relu")
        self.output_layer = layers.Dense(1, activation="sigmoid")

    def call(self, inputs):
        _x = self.layer1(inputs)
        _x = self.layer2(_x)
        _outputs = self.output_layer(_x)
        return _outputs
    
clModel = MyModel() 
```

### 학습 수행

```python
clModel.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = clModel.fit(x_train, y_train, epochs=500, batch_size=10)
```

### 데이터 검증

```python
score = clModel.evaluate(x_train, y_train, batch_size=10)
print(score)
```

### 데이터 예측

```python
y_predict = clModel.predict(x_new)
print(y_predict)
print(y_train[15:25])
```
